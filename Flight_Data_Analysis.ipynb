```python
# Colab cell 2
import pandas as pd
path = "/content/drive/MyDrive/Flight Data.csv"  # change if your path is different
df = pd.read_csv("/content/drive/MyDrive/Flight  Data.csv")
df.head(10)
```

```python
# Colab cell 3
import pandas as pd
import numpy as np

def clean_flight_df(df):
    df = df.copy()
    # Strip whitespace from string columns
    for c in df.select_dtypes(include=["object"]).columns:
        df[c] = df[c].astype(str).str.strip()

    # Standardize stops column (e.g. 'zero' -> 0, 'one' -> 1, 'non-stop' etc)
    def stops_to_int(x):
        x = str(x).lower().strip()
        map_ = {'zero':0, 'non-stop':0, 'non stop':0, 'nonstop':0, 'one':1, 'two':2, 'three':3}
        try:
            if x.isdigit():
                return int(x)
            return map_.get(x, np.nan)
        except:
            return np.nan
    if 'stops' in df.columns:
        df['stops_clean'] = df['stops'].apply(stops_to_int)
    else:
        df['stops_clean'] = np.nan

    # Price: remove non-numeric characters and convert to numeric
    if 'price' in df.columns:
        df['price'] = df['price'].astype(str).str.replace('[^0-9.]','', regex=True)
        df['price'] = pd.to_numeric(df['price'], errors='coerce')

    # Duration: try to convert to numeric hours (if format is like 2.17 which seems hours)
    if 'duration' in df.columns:
        df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

    # days_left
    if 'days_left' in df.columns:
        df['days_left'] = pd.to_numeric(df['days_left'], errors='coerce').astype('Int64')

    # create route column
    if 'source_city' in df.columns and 'destination_city' in df.columns:
        df['route'] = df['source_city'].astype(str) + " â†’ " + df['destination_city'].astype(str)

    # create departure time of day categories if departure_time is one of known labels
    def normalize_dep(x):
        x = str(x).lower()
        if 'early' in x: return 'Early Morning'
        if 'morning' in x: return 'Morning'
        if 'afternoon' in x: return 'Afternoon'
        if 'evening' in x: return 'Evening'
        if 'night' in x: return 'Night'
        return x.title()
    if 'departure_time' in df.columns:
        df['departure_period'] = df['departure_time'].apply(normalize_dep)
    if 'arrival_time' in df.columns:
        df['arrival_period'] = df['arrival_time'].apply(normalize_dep)

    # drop exact duplicates
    df = df.drop_duplicates()

    return df

df_clean = clean_flight_df(df)
df_clean.info()
df_clean.head()
```

```python
# Colab cell 4a - basic stats & missing values
print("Shape:", df_clean.shape)
display(df_clean.describe(include='all').transpose())
print("\nMissing values:")
display(df_clean.isna().sum())
```

```python
# Colab cell 4b - top airlines & routes
top_airlines = df_clean['airline'].value_counts().head(10)
top_routes = df_clean['route'].value_counts().head(10)
print("Top airlines:")
display(top_airlines)
print("\nTop routes:")
display(top_routes)
```

```python
# Colab cell 4c - price statistics
print("Price - basic stats")
display(df_clean['price'].describe())
print("Avg price by airline")
display(df_clean.groupby('airline')['price'].median().sort_values())
```

```python
# Colab cell 4d - flights by departure period and stops
display(pd.crosstab(df_clean['departure_period'], df_clean['stops_clean']))
display(df_clean.groupby('departure_period')['price'].median().sort_values(ascending=False))
```

```python
# Colab cell 4e - correlation like relationships
num_cols = ['price','duration','days_left','stops_clean']
display(df_clean[num_cols].corr())
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Flight Data Dashboard", layout="wide")

@st.cache_data
def load_data(path):
    return pd.read_csv(path)

DATA_PATH = "Flight_Data_cleaned.csv"  # ensure this file exists in the same folder

df = load_data(DATA_PATH)

# Sidebar filters
st.sidebar.header("Filters")
airlines = st.sidebar.multiselect("Airline", options=sorted(df['airline'].dropna().unique()), default=sorted(df['airline'].dropna().unique())[:5])
sources = st.sidebar.multiselect("Source City", options=sorted(df['source_city'].dropna().unique()), default=sorted(df['source_city'].dropna().unique()))
destinations = st.sidebar.multiselect("Destination City", options=sorted(df['destination_city'].dropna().unique()), default=sorted(df['destination_city'].dropna().unique()))
dep_periods = st.sidebar.multiselect("Departure Period", options=sorted(df['departure_period'].dropna().unique()), default=sorted(df['departure_period'].dropna().unique()))

# Apply filters
mask = df['airline'].isin(airlines) & df['source_city'].isin(sources) & df['destination_city'].isin(destinations) & df['departure_period'].isin(dep_periods)
df_f = df[mask].copy()

st.title("âœˆï¸ Flight Data Dashboard")
col1, col2, col3, col4 = st.columns(4)

# KPIs
with col1:
    st.metric("Total Flights", f"{len(df_f):,}")
with col2:
    st.metric("Median Price", f"â‚¹{int(df_f['price'].median() if not df_f['price'].isna().all() else 0):,}")
with col3:
    st.metric("Avg Duration (hrs)", round(float(df_f['duration'].median(skipna=True)),2) if 'duration' in df_f else "N/A")
with col4:
    st.metric("Unique Routes", f"{df_f['route'].nunique():,}")

st.markdown("---")

# Charts: flights by airline
st.subheader("Flights by Airline")
air_chart = df_f['airline'].value_counts().reset_index()
air_chart.columns = ['airline','count']
fig1 = px.bar(air_chart, x='airline', y='count', title="Flights per Airline", labels={'count':'Number of Flights'})
st.plotly_chart(fig1, use_container_width=True)

# Top routes & price distribution
colA, colB = st.columns([2,1])
with colA:
    st.subheader("Top Routes")
    top_routes = df_f['route'].value_counts().reset_index().head(20)
    top_routes.columns = ['route','count']
    fig2 = px.bar(top_routes, x='route', y='count', title="Top Routes (source â†’ dest)", labels={'count':'Number of Flights'})
    fig2.update_layout(xaxis={'categoryorder':'total descending'})
    st.plotly_chart(fig2, use_container_width=True)
with colB:
    st.subheader("Price Distribution")
    fig3 = px.histogram(df_f, x='price', nbins=40, title="Price Distribution")
    st.plotly_chart(fig3, use_container_width=True)

# Median price by airline
st.subheader("Median Price by Airline")
price_by_air = df_f.groupby('airline')['price'].median().reset_index().sort_values('price', ascending=False)
fig4 = px.bar(price_by_air, x='airline', y='price', title="Median Price by Airline", labels={'price':'Median Price'})
st.plotly_chart(fig4, use_container_width=True)

# Departure period heatmap of counts vs stops
st.subheader("Departure Period vs Stops")
if 'departure_period' in df_f.columns and 'stops_clean' in df_f.columns:
    heat = pd.crosstab(df_f['departure_period'], df_f['stops_clean'])
    fig5 = px.imshow(heat, labels=dict(x="Stops", y="Departure Period", color="Count"), text_auto=True)
    st.plotly_chart(fig5, use_container_width=True)

# Data table and download
st.subheader("Filtered Data (sample)")
st.dataframe(df_f.head(200))

@st.cache_data
def convert_df_to_csv(df):
    return df.to_csv(index=False).encode('utf-8')

csv = convert_df_to_csv(df_f)
st.download_button("Download filtered data as CSV", data=csv, file_name="filtered_flights.csv", mime="text/csv")
```
